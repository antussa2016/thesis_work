```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('thesis_work/dataset/output.csv')
```

```python
df.head(1)
```

```python
columns_to_drop = ['stackoverflow_id','question_count','answer_count','reputation','answer_tag','question_tag','preferred_tags1', 'preferred_tags2','preferred_tags3']
df.drop(columns_to_drop, axis=1, inplace=True)
```

```python
df.head(1)
```

```python
df_melted = df.melt(id_vars=['twitter_id','Openness', 'Conscientiousness','Extraversion','Agreeableness','Neuroticism'],
                    value_vars=['micro-oop','oop','web-pro','database','mobile','statistic','other'],
                    var_name='Class_Label',
                    value_name='Value')
```

```python
df_filtered = df_melted[df_melted['Value'] != 0]
```

```python
df_filtered.drop(columns=['Value'], inplace=True)
```

```python

df_filtered.sort_values(by='twitter_id', inplace=True)
```

```python
df_filtered.reset_index(drop=True, inplace=True)
```

```python
new_df = df_filtered.reset_index(drop=True).copy()
```
